```python
url="/content/sample_data/IRIS.csv"
import pandas as pd
dataset=pd.read_csv(url)
```

```python
dataset.head(10)
```

```python
dataset['species'].unique()
```

```python
dataset.shape
```

```python
dataset['species'].value_counts()
```

```python
tmp=list()
for etiqueta in dataset['species']:
  if etiqueta == 'Iris-setosa':
    tmp.append(1)
  elif etiqueta == 'Iris-versicolor':
      tmp.append(2)
  else:
        tmp.append(3)

dataset['etiqueta']=tmp
```

```python
dataset.head(10)
```

```python
#array(['Iris-setosa'.'Iris-versicolor'.'Iris-virginica'].dtype=object)
maskSetosa=dataset['species']=='Iris-setosa'
maskVersicolor=dataset['species']=='Iris-versicolor'
maskVirginica=dataset['species']=='Iris-virginica'



setosa=dataset[maskSetosa]
versicolor=dataset[maskVersicolor]
virginica=dataset[maskVirginica]
```

```python
setosa.head(3)
```

```python
print(setosa.shape)
print(versicolor.shape)
print(virginica.shape)
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
fig.set_size_inches(14,6)

ax.scatter(setosa['sepal_length'],setosa['petal_width'],label='setosa')
ax.scatter(versicolor['sepal_length'],versicolor['petal_width'],label='versicolor')
ax.scatter(virginica['sepal_length'],virginica['petal_width'],label='virginica')

ax.legend()
```
